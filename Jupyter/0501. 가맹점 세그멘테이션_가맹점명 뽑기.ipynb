```python
data = pd.read_csv("../Result/0500. 가맹점 세그멘테이션/가맹점군집.csv", encoding='cp949')
```

```python
sum = data.groupby(['연도', '가맹점 주소', '골목상권분류', 'kmeans_clusters', '대분류', '가맹점명'])['총 매출'].sum().reset_index()
sum
```

```python
store = sum.groupby(['연도', '가맹점 주소', '골목상권분류', 'kmeans_clusters', '대분류']).apply(lambda x: x.nlargest(2, '총 매출')).reset_index(drop=True)
```

```python
store.to_csv("../Result/0500. 가맹점 세그멘테이션/가맹점명.csv", index=False, encoding='cp949')
```
